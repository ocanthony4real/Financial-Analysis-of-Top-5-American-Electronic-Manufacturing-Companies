# Financial Analysis of Top 5 American Electronic Manufacturing Companies

## Brief Note
Below are the names of the top 5 electronic manufacturing companies in America as suggested by http://www.electronicsandyou.com/blog/top-electronic-companies-in-usa.html (as at 15/8/2022):

1) Cisco Systems (CSCO)

2) Texas Instruments (TXN)

3) IBM (IBM)

4) Intel (INTC)

5) HP (HPE)

For easy identification, their tickers, as used in Yahoo Finance, are written in paranthesis and will be a mode of identification throughout the analysis. For each company, the balance sheet, financials sheet, and cashflow sheet will be exctracted from Yahoo Finance before analysis will be carried out on them. 

The target audience for this analysis are potential investors who wish to get an indepth comparisons of the financial performances of these companies in the financial market.

Sections in this analysis include data exctraction, data cleaning, and descriptive analysis. 

Please note, for the sake of this analaysis, only the year of the data retrieved is regarded (i.e the month and date of the data is discarded). This is to ensure uniformity among the data of the five companies.  This may not be ideal in real life situations.

## Data Extraction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import yfinance as yf
```

```python
CSCO = yf.Ticker('CSCO')
TXN = yf.Ticker('TXN')
IBM = yf.Ticker('IBM')
INTC = yf.Ticker('INTC')
HPE = yf.Ticker('HPE')

#Assigne tickers for each company
```

### Data extraction and wrangling of balance sheets

```python
CSCO_balance_sheet = CSCO.balance_sheet.transpose()   #scrap balance sheet and transpose it
CSCO_balance_sheet.index = CSCO_balance_sheet.index.year    #change the index from date type to year
CSCO_balance_sheet.index.names = ['Year']    #rename the index to 'year'
CSCO_balance_sheet = CSCO_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top 

TXN_balance_sheet = TXN.balance_sheet.transpose()    #scrap balance sheet and transpose it
TXN_balance_sheet.index = TXN_balance_sheet.index.year    #change the index from date type to year
TXN_balance_sheet.index.names = ['Year']     #rename the index to 'year'
TXN_balance_sheet = TXN_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top

IBM_balance_sheet = IBM.balance_sheet.transpose()    #scrap balance sheet and transpose it
IBM_balance_sheet.index = IBM_balance_sheet.index.year    #change the index from date type to year
IBM_balance_sheet.index.names = ['Year']     #rename the index to 'year'
IBM_balance_sheet = IBM_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top

INTC_balance_sheet = INTC.balance_sheet.transpose()    #scrap balance sheet and transpose it
INTC_balance_sheet.index = INTC_balance_sheet.index.year    #change the index from date type to year
INTC_balance_sheet.index.names = ['Year']     #rename the index to 'year'
INTC_balance_sheet = INTC_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top

HPE_balance_sheet = HPE.balance_sheet.transpose()    #scrap balance sheet and transpose it
HPE_balance_sheet.index = HPE_balance_sheet.index.year    #change the index from date type to year
HPE_balance_sheet.index.names = ['Year']     #rename the index to 'year'
HPE_balance_sheet = HPE_balance_sheet.loc[::-1]    #reverse the ordering of the table from bottom to top
```

```python
comp_list = ['CSCO', 'TXN', 'IBM', 'INTC', 'HPE']    #list of the tickers
```

```python
total_balance_sheets = pd.concat([CSCO_balance_sheet, TXN_balance_sheet, IBM_balance_sheet, INTC_balance_sheet, HPE_balance_sheet], join = 'outer', axis = 1, keys = comp_list)
total_balance_sheets
#concatenate the five sheets extracted
```

```python
total_balance_sheets.isnull().all().value_counts() #check for empty columns
```

### Data extraction and wrangling of financials sheets

```python
CSCO_financials = CSCO.financials.transpose()    #scrap balance sheet and transpose it
CSCO_financials.index = CSCO_financials.index.year    #change the index from date type to year
CSCO_balance_sheet.index.names = ['Year']     #rename the index to 'year'
CSCO_financials = CSCO_financials.loc[::-1]    #reverse the ordering of the table from bottom to top

TXN_financials = TXN.financials.transpose()    #scrap balance sheet and transpose it
TXN_financials.index = TXN_financials.index.year    #change the index from date type to year
TXN_financials.index.names = ['Year']     #rename the index to 'year'
TXN_financials = TXN_financials.loc[::-1]    #reverse the ordering of the table from bottom to top

IBM_financials = IBM.financials.transpose()    #scrap balance sheet and transpose it
IBM_financials.index = IBM_financials.index.year    #change the index from date type to year
IBM_financials.index.names = ['Year']     #rename the index to 'year'
IBM_financials = IBM_financials.loc[::-1]    #reverse the ordering of the table from bottom to top

INTC_financials = INTC.financials.transpose()    #scrap balance sheet and transpose it
INTC_financials.index = INTC_financials.index.year    #change the index from date type to year
INTC_financials.index.names = ['Year']     #rename the index to 'year'
INTC_financials = INTC_financials.loc[::-1]    #reverse the ordering of the table from bottom to top

HPE_financials = HPE.financials.transpose()    #scrap balance sheet and transpose it
HPE_financials.index = HPE_financials.index.year    #change the index from date type to year
HPE_financials.index.names = ['Year']     #rename the index to 'year'
HPE_financials = HPE_financials.loc[::-1]    #reverse the ordering of the table from bottom to top
```

```python
total_financials_sheets = pd.concat([CSCO_financials, TXN_financials, IBM_financials, INTC_financials, HPE_financials], join = 'outer', axis = 1, keys = comp_list)
total_financials_sheets
#concatenate the five sheets extracted
```

```python
total_financials_sheets.isnull().all().value_counts() #there are 28 empty columns
```

```python
total_financials_sheets.dropna(how = 'all', axis = 1, inplace = True) #drop the empty columns
```

### Data extraction and wrangling of cashflow sheets

```python
CSCO_cashflow = CSCO.cashflow.transpose()    #scrap balance sheet and transpose it
CSCO_cashflow.index = CSCO_cashflow.index.year    #change the index from date type to year
CSCO_cashflow.index.names = ['Year']     #rename the index to 'year'
CSCO_cashflow = CSCO_cashflow.loc[::-1]    #reverse the ordering of the table from bottom to top

TXN_cashflow = TXN.cashflow.transpose()    #scrap balance sheet and transpose it
TXN_cashflow.index = TXN_cashflow.index.year    #change the index from date type to year
TXN_cashflow.index.names = ['Year']     #rename the index to 'year'
TXN_cashflow = TXN_cashflow.loc[::-1]    #reverse the ordering of the table from bottom to top

IBM_cashflow = IBM.cashflow.transpose()    #scrap balance sheet and transpose it
IBM_cashflow.index = IBM_cashflow.index.year    #change the index from date type to year
IBM_cashflow.index.names = ['Year']     #rename the index to 'year'
IBM_cashflow = IBM_cashflow.loc[::-1]    #reverse the ordering of the table from bottom to top

INTC_cashflow = INTC.cashflow.transpose()    #scrap balance sheet and transpose it
INTC_cashflow.index = INTC_cashflow.index.year    #change the index from date type to year
INTC_cashflow.index.names = ['Year']     #rename the index to 'year'
INTC_cashflow = INTC_cashflow.loc[::-1]    #reverse the ordering of the table from bottom to top

HPE_cashflow = HPE.cashflow.transpose()    #scrap balance sheet and transpose it
HPE_cashflow.index = HPE_cashflow.index.year    #change the index from date type to year
HPE_cashflow.index.names = ['Year']     #rename the index to 'year'
HPE_cashflow = HPE_cashflow.loc[::-1]    #reverse the ordering of the table from bottom to top
```

```python
total_cashflow_sheets = pd.concat([CSCO_cashflow, TXN_cashflow, IBM_cashflow, INTC_cashflow, HPE_cashflow], join = 'outer', axis = 1, keys = comp_list)
total_cashflow_sheets
#concatenate the five sheets extracted
```

```python
total_cashflow_sheets.isnull().all().value_counts()    #there are no empty columns
```

## Descriptive analysis and possible recommendations

#### Depreciation

```python
Deprec = pd.DataFrame() #create empty data frame
```

```python
for tick in comp_list:
    Deprec[tick + ' Deprec'] = total_cashflow_sheets[tick]['Depreciation']

Deprec

#create Depreciation table
```

```python
v = Deprec.iloc[0].to_numpy()
w = Deprec.iloc[1].to_numpy()
x = Deprec.iloc[2].to_numpy()
y = Deprec.iloc[3].to_numpy()
#assign rows of the data frame (deprec)

fig, ax = plt.subplots( figsize=(8,4))
#create empty figure

ax.barh(comp_list, v, label = '2018', height = 0.6)
ax.barh(comp_list, w, left = v, label = '2019', height = 0.6)
ax.barh(comp_list, x, left = v+w, label = '2020', height = 0.6)
ax.barh(comp_list, y, left = v+w+x, label = '2021', height = 0.6)

ax.xaxis.grid(linestyle = '--')    #set vertical line style for the x axis

ax.set_ylabel("Companies", fontsize = 12, color = 'dimgray')

ax.set_xlabel("Depreciation", fontsize = 13, color = 'dimgray')

ax.set_title("Cummulative Depreciation", fontsize = 15, color = 'k')


for pos in ['right', 'top']:
    ax.spines[pos].set_visible(False)
#remove the right and top spines
    
for pos in ['left', 'bottom']:
    ax.spines[pos].set_color('lightgrey')
#change color of left and bottom spines
    
for pos in ['x', 'y']:
    ax.tick_params(pos, colors = 'dimgray')
#change color of the x and y ticks

ax.legend(loc = (1.05, 0.72))

plt.show()
```

INTC's assets has experienced a relatively high level of depreciation over the past four years, unlike TXN which has kept it's depreciation minimal. This means that investments in INTC will yield smaller return on equity for shareholdders when compared to that of other companies.

#### Net Profit Margin

```python
Net_Profit_Margin = pd.DataFrame()    #create empty data frame
```

```python
for tick in comp_list:
    Net_Profit_Margin[tick +' NPM'] = total_financials_sheets[tick]['Net Income']/total_financials_sheets[tick]['Total Revenue']
    
Net_Profit_Margin.index.names= ['Year'] 
Net_Profit_Margin

#create Net Profit Margin table
```

```python
fig, ax = plt.subplots(figsize=(8,4))
#create empty figure

plt.xticks([2018, 2019, 2020, 2021])
#set values of x ticks

ax.plot(Net_Profit_Margin.index, 'CSCO NPM', data = Net_Profit_Margin, alpha = 0.4)
ax.plot(Net_Profit_Margin.index, 'TXN NPM', data = Net_Profit_Margin, color = 'maroon', linestyle = '-.')
ax.plot(Net_Profit_Margin.index, 'IBM NPM', data = Net_Profit_Margin, alpha = 0.5)
ax.plot(Net_Profit_Margin.index, 'INTC NPM', data = Net_Profit_Margin, alpha = 0.6)
ax.plot(Net_Profit_Margin.index, 'HPE NPM', data = Net_Profit_Margin, alpha = 0.4)

ax.set_ylabel("Net Profit Margin (NPM)", fontsize = 12, color = 'dimgray')

ax.set_xlabel("Years", fontsize = 13, color = 'dimgray')

ax.set_title("Yearly Net Profit Margin", fontsize = 15, color = 'k')

for pos in ['right', 'top']:
    ax.spines[pos].set_visible(False)
#remove the right and top spines

for pos in ['left', 'bottom']:
    ax.spines[pos].set_color('lightgrey')
#change color of left and bottom spines

for pos in ['x', 'y']:
    ax.tick_params(pos, colors = 'dimgray')
#change color of the x and y ticks

ax.legend(bbox_to_anchor = (1.3, 1))
plt.show()
```

TXN is doing very well in converting its sales to profit. Their net profit margin has also been on a steady increase since 2019, unlike INTC and IBM which have been experiencing a decline in their net profit margin. TXN is definitely a company that potential investors should watch out for when they want to invest in electronic manufacturing companies. 

#### Capital Expenditure Ratio (CAPEX)

```python
CAPEX_Ratio = pd.DataFrame()    #create empty data frame
```

```python
for tick in comp_list:
    CAPEX_Ratio[tick +' CAPEX_Ratio'] = (total_cashflow_sheets[tick]['Capital Expenditures'].abs())/total_financials_sheets[tick]['Total Revenue']

CAPEX_Ratio

#create CAPEX ratio table
```

```python
fig, ax = plt.subplots(figsize=(8,4))
#create empty figure

plt.xticks([2018, 2019, 2020, 2021])
#set values of x ticks

ax.plot(CAPEX_Ratio.index, 'CSCO CAPEX_Ratio', data = CAPEX_Ratio, alpha = 0.4)
ax.plot(CAPEX_Ratio.index, 'TXN CAPEX_Ratio', data = CAPEX_Ratio, color = 'maroon', alpha = 0.4)
ax.plot(CAPEX_Ratio.index, 'IBM CAPEX_Ratio', data = CAPEX_Ratio, alpha = 0.4)
ax.plot(CAPEX_Ratio.index, 'INTC CAPEX_Ratio',  linestyle = '-.', data = CAPEX_Ratio)
ax.plot(CAPEX_Ratio.index, 'HPE CAPEX_Ratio', data = CAPEX_Ratio, alpha = 0.4)

ax.set_ylabel("Ratio", fontsize = 12, color = 'dimgray')

ax.set_xlabel("Years", fontsize = 13, color = 'dimgray')

ax.set_title("Yearly CAPEX Ratio", fontsize=14, color='k')

for pos in ['right', 'top']:
    ax.spines[pos].set_visible(False)
#remove the right and top spines

for pos in ['left', 'bottom']:
    ax.spines[pos].set_color('lightgrey')
#change color of left and bottom spines

for pos in ['x', 'y']:
    ax.tick_params(pos, colors = 'dimgray')
#change color of the x and y ticks

ax.legend(bbox_to_anchor = (1.4, 1))
plt.show()
```

INTC has been investing heavily, more than their counterparts in their capital expenditure. In the future, they are likely
to enjoy high financial turnovers. This is a positive indication for investors.

HPE's working capital ratio has been negative over the past few years. The means that they barely have enough

#### Return on Equity

```python
ROE = pd.DataFrame()    #create empty data frame
```

```python
ROE_list= ['CSCO ROE', 'TXN ROE', 'IBM ROE', 'INTC ROE', 'HPE ROE']
```

```python
for tick in comp_list:
    
    ROE[tick +' Net Worth'] = total_balance_sheets[tick]['Total Assets'] - total_balance_sheets[tick]['Total Liab']
    ROE[tick+' ROE'] = total_cashflow_sheets[tick]['Net Income']/ROE[tick+' Net Worth']
    
Return_on_Equity = ROE.loc[:, ROE_list]
Return_on_Equity

#create Return on Equity table
```

```python
fig, ax = plt.subplots(figsize=(8,4))
#create empty figure

plt.xticks([2018, 2019, 2020, 2021])
#set values of x ticks

ax.plot(Return_on_Equity.index, 'CSCO ROE', data = Return_on_Equity, alpha = 0.4)
ax.plot(Return_on_Equity.index, 'TXN ROE', data = Return_on_Equity, color = 'maroon', linestyle = '-.')
ax.plot(Return_on_Equity.index, 'IBM ROE', data = Return_on_Equity, alpha = 0.4)
ax.plot(Return_on_Equity.index, 'INTC ROE', data = Return_on_Equity, alpha = 0.4)
ax.plot(Return_on_Equity.index, 'HPE ROE', data = Return_on_Equity, alpha = 0.4)

ax.set_ylabel("Return_on_Equity", fontsize = 12, color = 'dimgray')

ax.set_xlabel("Years", fontsize = 13, color = 'dimgray')

ax.set_title("Yearly Return on Equity", fontsize=14, color='k')

for pos in ['right', 'top']:
    ax.spines[pos].set_visible(False)
#remove the right and top spines

for pos in ['left', 'bottom']:
    ax.spines[pos].set_color('lightgrey')
#change color of left and bottom spines 

for pos in ['x', 'y']:
    ax.tick_params(pos, colors = 'dimgray')
#change color of the x and y ticks

ax.legend(bbox_to_anchor = (1.4, 1))
plt.show()
```

From the information on the chart, TXN has a high and stable return on equity. Investors are more likely to get better returns
from their investment in the company more than their investments in other conpanies.
