```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import yfinance as yf
#import datetime
```

```python
CSCO = yf.Ticker('CSCO')
TXN = yf.Ticker('TXN')
IBM = yf.Ticker('IBM')
INTC = yf.Ticker('INTC')
HPE = yf.Ticker('HPE')
```

```python
CSCO_balance_sheet = CSCO.balance_sheet.transpose()
CSCO_balance_sheet.index = CSCO_balance_sheet.index.year
CSCO_balance_sheet.index.names = ['Year']
CSCO_balance_sheet = CSCO_balance_sheet.loc[::-1]

TXN_balance_sheet = TXN.balance_sheet.transpose()
TXN_balance_sheet.index = TXN_balance_sheet.index.year
TXN_balance_sheet.index.names = ['Year']
TXN_balance_sheet = TXN_balance_sheet.loc[::-1]


IBM_balance_sheet = IBM.balance_sheet.transpose()
IBM_balance_sheet.index = IBM_balance_sheet.index.year
IBM_balance_sheet.index.names = ['Year']
IBM_balance_sheet = IBM_balance_sheet.loc[::-1]

INTC_balance_sheet = INTC.balance_sheet.transpose()
INTC_balance_sheet.index = INTC_balance_sheet.index.year
INTC_balance_sheet.index.names = ['Year']
INTC_balance_sheet = INTC_balance_sheet.loc[::-1]

HPE_balance_sheet = HPE.balance_sheet.transpose()
HPE_balance_sheet.index = HPE_balance_sheet.index.year
HPE_balance_sheet.index.names = ['Year']
HPE_balance_sheet = HPE_balance_sheet.loc[::-1]
```

```python
comp_list = ['CSCO', 'TXN', 'IBM', 'INTC', 'HPE']
```

```python
total_balance_sheets = pd.concat([CSCO_balance_sheet, TXN_balance_sheet, IBM_balance_sheet, INTC_balance_sheet, HPE_balance_sheet], join = 'outer', axis = 1, keys = comp_list)
total_balance_sheets
```

```python
CSCO_financials = CSCO.financials.transpose()
CSCO_financials.index = CSCO_financials.index.year
CSCO_balance_sheet.index.names = ['Year']
CSCO_financials = CSCO_financials.loc[::-1]

TXN_financials = TXN.financials.transpose()
TXN_financials.index = TXN_financials.index.year
TXN_financials.index.names = ['Year']
TXN_financials = TXN_financials.loc[::-1]

IBM_financials = IBM.financials.transpose()
IBM_financials.index = IBM_financials.index.year
IBM_financials.index.names = ['Year']
IBM_financials = IBM_financials.loc[::-1]

INTC_financials = INTC.financials.transpose()
INTC_financials.index = INTC_financials.index.year
INTC_financials.index.names = ['Year']
INTC_financials = INTC_financials.loc[::-1]

HPE_financials = HPE.financials.transpose()
HPE_financials.index = HPE_financials.index.year
HPE_financials.index.names = ['Year']
HPE_financials = HPE_financials.loc[::-1]
```

```python
total_financials_sheets = pd.concat([CSCO_financials, TXN_financials, IBM_financials, INTC_financials, HPE_financials], join = 'outer', axis = 1, keys = comp_list)
total_financials_sheets['CSCO']
```

```python
CSCO_cashflow = CSCO.cashflow.transpose()
CSCO_cashflow.index = CSCO_cashflow.index.year
CSCO_cashflow.index.names = ['Year']
CSCO_cashflow = CSCO_cashflow.loc[::-1]

TXN_cashflow = TXN.cashflow.transpose()
TXN_cashflow.index = TXN_cashflow.index.year
TXN_cashflow.index.names = ['Year']
TXN_cashflow = TXN_cashflow.loc[::-1]

IBM_cashflow = IBM.cashflow.transpose()
IBM_cashflow.index = IBM_cashflow.index.year
IBM_cashflow.index.names = ['Year']
IBM_cashflow = IBM_cashflow.loc[::-1]

INTC_cashflow = INTC.cashflow.transpose()
INTC_cashflow.index = INTC_cashflow.index.year
INTC_cashflow.index.names = ['Year']
INTC_cashflow = INTC_cashflow.loc[::-1]

HPE_cashflow = HPE.cashflow.transpose()
HPE_cashflow.index = HPE_cashflow.index.year
HPE_cashflow.index.names = ['Year']
HPE_cashflow = HPE_cashflow.loc[::-1]
```

```python
total_cashflow_sheets = pd.concat([CSCO_cashflow, TXN_cashflow, IBM_cashflow, INTC_cashflow, HPE_cashflow], join = 'outer', axis = 1, keys = comp_list)
total_cashflow_sheets
```

```python
Net_Profit_Margin = pd.DataFrame()
```

```python

for tick in comp_list:
    Net_Profit_Margin[tick +' NPM'] = total_financials_sheets[tick]['Net Income']/total_financials_sheets[tick]['Total Revenue']
Net_Profit_Margin.index.names= ['Year']    
Net_Profit_Margin
```

```python
sns.lineplot(x = 'Year', y = 'CSCO NPM', data = Net_Profit_Margin, color = 'red')
sns.lineplot(x = 'Year', y = 'TXN NPM', data = Net_Profit_Margin, color = 'black')
sns.lineplot(x = 'Year', y = 'IBM NPM', data = Net_Profit_Margin, color = 'black')
sns.lineplot(x = 'Year', y = 'INTC NPM', data = Net_Profit_Margin, color = 'black')
sns.lineplot(x = 'Year', y = 'HPE NPM', data = Net_Profit_Margin, color = 'black')
```

```python
CAPEX_Ratio = pd.DataFrame()
```

```python
for tick in comp_list:
    CAPEX_Ratio[tick +' CAPEX_Ratio'] = total_cashflow_sheets[tick]['Capital Expenditures']/total_financials_sheets[tick]['Total Revenue']
CAPEX_Ratio
```

```python
OER = pd.DataFrame()
```

```python
for tick in comp_list:
    OER[tick + ' OER'] = total_financials_sheets[tick]['Total Operating Expenses']/total_financials_sheets[tick]['Total Revenue']
OER.index.names= ['Year']
OER
```

```python
Working_Capital_Ratio = pd.DataFrame()
```

```python
for tick in comp_list:
    Working_Capital_Ratio[tick +' WCR'] = total_balance_sheets[tick]['Total Current Assets'] - total_balance_sheets[tick]['Total Current Liabilities']
Working_Capital_Ratio
```

```python
Debt_to_Equity = pd.DataFrame()
```

```python
for tick in comp_list:
    Debt_to_Equity[tick +' D/E'] = total_balance_sheets[tick]['Total Liab']/ total_balance_sheets[tick]['Total Stockholder Equity']
Debt_to_Equity
```

```python
#Return_on_Equity = pd.DataFrame(columns = [comp_list, [' Net Worth']*5])
#pd.MultiIndex.from_frame(Return_on_Equity)
ROE = pd.DataFrame()
```

```python
ROE_list= ['CSCO ROE', 'TXN ROE', 'IBM ROE', 'INTC ROE', 'HPE ROE']
```

```python
for tick in comp_list:
    ROE[tick +' Net Worth'] = total_balance_sheets[tick]['Total Assets'] - total_balance_sheets[tick]['Total Liab']
    ROE[tick+' ROE'] = total_cashflow_sheets[tick]['Net Income']/ROE[tick+' Net Worth']
Return_on_Equity = ROE.loc[:, ROE_list]
Return_on_Equity
```
